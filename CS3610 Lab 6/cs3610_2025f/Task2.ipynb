```python
from typing import Type, Self 
from abc import ABC, abstractmethod
from operator import methodcaller
```

```python
class Product(ABC):
    def Get_Price():
        pass
    def get_Discription():
        pass
```

```python
class BurgerClass(Product):
    def __init__(self):
        self.Price = 0
        self.calories = 0
        self.description = ""
    
    def Get_Price(self):
        return self.Price
    def get_Discription(self):
        return self.description

class PizzaClass(Product):
    def __init__(self):
        self.Price = 0
        self.calories = 0
        self.description = ""
        self.size = "small"

    def Get_Price(self):
        return self.Price
    def get_Discription(self):
        return self.description
    
class NoodlesClass(Product):
    def __init__(self):
        self.Price = 0
        self.calories = 0
        self.description = ""
    
    def Get_Price(self):
        return self.Price
    def get_Discription(self):
        return self.description

class CutletClass(Product):
    def __init__(self):
        self.Price = 0
        self.calories = 0
        self.description = ""
        self.size = "small"

    def Get_Price(self):
        return self.Price
    def get_Discription(self):
        return self.description
```

```python
class VegBurger(BurgerClass):
    def __init__(self):
        self.Price = 12
        self.calories = 800
        self.description = "A Burger with a Veg patty, Lettuce, Tomatos, Ketchup, Mustard, and Onions"

    def Get_Price(self):
        print(self.Price)
    def get_Discription(self):
        print(self.description)
        

class Burger(BurgerClass):
    def __init__(self):
        self.Price = 10
        self.calories = 1000
        self.description = "A Burger with a Meat patty, Onions, BQQ sause, Mushrooms, and Bacon"

    def Get_Price(self):
        print(self.Price)
    def get_Discription(self):
        print(self.description)

class VegPizza(PizzaClass):
    def __init__(self):
        self.Price = 11
        self.calories = 700
        self.description = "A pizza with Pepper's, Spinich, onions, and tomatos on top"
        self.size = "Small"

    def Get_Price(self):
        print(self.Price)
    def get_Discription(self):
        print(self.description)

class Pizza(PizzaClass):
    def __init__(self):
        self.Price = 11
        self.calories = 950
        self.description = "A Pizza with Pepperoni, Bacon, and Mushrooms"
        self.size = "Medium"

    def Get_Price(self):
        print(self.Price)
    def get_Discription(self):
        print(self.description)

class VegNoodles(NoodlesClass):
    def __init__(self):
        self.Price = 12
        self.calories = 800
        self.description = "Noodles using a Veg Brooth"

    def Get_Price(self):
        print(self.Price)
    def get_Discription(self):
        print(self.description)
        

class Noodles(NoodlesClass):
    def __init__(self):
        self.Price = 10
        self.calories = 1000
        self.description = "Noodles using a Meat Brooth"

    def Get_Price(self):
        print(self.Price)
    def get_Discription(self):
        print(self.description)

class Cutlet(CutletClass):
    def __init__(self):
        self.Price = 11
        self.calories = 950
        self.description = "A Meat Cutlet"

    def Get_Price(self):
        print(self.Price)
    def get_Discription(self):
        print(self.description)
```

```python

class ProductFactory(ABC):
    def __init__():
        super().__init__()
        
    @abstractmethod
    def createBurger(self) -> BurgerClass:
        pass
    
    def createPizza(self) -> PizzaClass:
        pass

    def createNoodles(self) -> Noodles:
        pass
    
    def createCutlet(self) -> Cutlet:
        pass

```

```python

class VegProductFactory(ProductFactory):

    def createBurger(self) -> VegBurger:
        print("I am making the Burger...")
        return VegBurger()
    
    def createPizza(self) -> VegPizza:
        print("I am making the Pizza...")
        return VegPizza()
    
    def createNoodles(self) -> Noodles:
        print("I am making the Noodes...")
        return Noodles()
```

```python

class NonVegProductFactory(ProductFactory):

    def createBurger(self) -> Burger:
        print("I am making the Burger...")
        return Burger()
    
    def createPizza(self) -> Pizza:
        print("I am making the Pizza...")
        return Pizza()
    
    def createNoodles(self) -> Noodles:
        print("I am making the Noodes...")
        return Noodles()
    
    def createCutlet(self) -> Cutlet:
        print("I am making the Cutlet...")
        return Cutlet()
```

```python
class FoodApp(ABC):
    def __init__(self):
        self.factory = {"nv": NonVegProductFactory, "v": VegProductFactory}
    #a static method of a class that returns an object of that class' type. 
    #But unlike a constructor, the actual object it returns might be an instance of a subclass. 
    #!!! Unlike a constructor, an existing object might be reused, instead of a new object created.

    def MakeOrder(self, items: list[str]) -> list[Product]:
        foodList: Product = []

        for i in items:
            if "veg" in i.lower():
                currentFac:ProductFactory = self.factory.get("v")
            else:
                currentFac:ProductFactory = self.factory.get("nv")
            
            if "burger" in i.lower():
                foodList.append(currentFac.createBurger(self)) 
            elif "pizza" in i.lower():
                foodList.append(currentFac.createPizza(self)) 
            elif "noodles" in i.lower():
                foodList.append(currentFac.createNoodles(self))  
            elif "cutlet" in i.lower():
                foodList.append(currentFac.createCutlet(self))

        return foodList       



           

    def GetOrderDescription(self, orders:list[Product]) -> str: #An abstract interface method

        for F in orders:
            F.get_Discription()
```

```python
Fa = FoodApp()
Order = Fa.MakeOrder(['VegPizza', 'Burger', 'Cutlet'])
Fa.GetOrderDescription(Order)
```
