```python
class Product(ABC):
    def Get_Price():
        pass
    def get_Discription():
        pass
```

```python
class Burger(Product):
    def __init__(self):
        self.Price = 0
        self.calories = 0
        self.description = ""
    
    def Get_Price(self):
        return self.Price
    def get_Discription(self):
        return self.description

class Pizza(Product):
    def __init__(self):
        self.Price = 0
        self.calories = 0
        self.description = ""
        self.size = ""

    def Get_Price(self):
        return self.Price
    def get_Discription(self):
        return self.description
```
