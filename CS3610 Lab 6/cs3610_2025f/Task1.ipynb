```python
from typing import Type, Self 
from abc import ABC, abstractmethod
from operator import methodcaller

class Zoo(ABC):
    def __init__(self):
        self.Animal_List: list[Type[Animal]]=[]
        self.Itinerary: str
    
    @staticmethod #a static method of a class that returns an object of that class' type. 
    #But unlike a constructor, the actual object it returns might be an instance of a subclass. 
    #!!! Unlike a constructor, an existing object might be reused, instead of a new object created.
    @abstractmethod
    def create_animals(): #An abstract interface method
       pass

    def List_Animals(self):
        print("list of animals: ")
        for A in self.animal:
            print(A)
        
    def AskEachAnimal(self):
        #print(len(self.Animal_List))
        for animal in self.Animal_List:
            print (f"{animal} says: ", end='')
            animal.Say()

    def StartTour(self):
        self.create_animals()
        print(self.Itinerary)
        self.List_Animals()
        self.AskEachAnimal()
    
```

```python
class Toronto_Zoo(Zoo):   
    def __init__(self):
        super().__init__()
        self.Itinerary = "Welcome to the Toronto Zoo! Our Itinerary: 1) African Savanna. 2) Tundra Trek"
        self.animal = ['Lion','Elephant','Penguin','Polar_Bear']
   
    def create_animals(self):
        #animal=['Lion','Elephant','Penguin','Polar Bear']
        animals=list(map(Animal_Factory.create_animals,self.animal))
        
        #print(len(animals))
        for animal in animals:
            if animal:
                self.Animal_List.append(animal)  


class Calgary_Zoo(Zoo):
    def __init__(self):
        super().__init__()
        self.Itinerary = "Welcome to the Toronto Zoo! Our Itinerary: 1) Wild Canada 2) Penguin Plunge."
        self.animal = ['Grizzly_Bear','Moose','Penguin']

    def create_animals(self):
        #animal=['Grizzly Bear','Moose','Penguin']
        animals=list(map(Animal_Factory.create_animals,self.animal))
        
        for animal in animals:
            if animal:
                self.Animal_List.append(animal) 
```

```python
class Animal(ABC):
    @staticmethod
    @abstractmethod
    def Say(): #A static interface method
        pass
```

```python
Animals={'Elephant': Toronto_Zoo, 'Lion': Toronto_Zoo, 'Penguin': Toronto_Zoo, 'Polar_Bear': Toronto_Zoo,'Penquin': Calgary_Zoo, 'Moose': Calgary_Zoo,'Grizzly_Bear': Calgary_Zoo}

class Animal_Factory:
    '''The Factory Method Class'''

    @staticmethod
    def create_animals(objType: str) -> Animal: #A static method to get a concrete product
        try:
            if objType in Animals:
                return eval(objType)()
            else:
                raise Exception("I can't make this object")
        except Exception as _e:
            print(_e)
        return None

    @abstractmethod
    def Say():
        pass
```

```python
class Lion(Animal): 
    def __str__(self):
        return"Lion"

    def Say(self):
        print("I Am a Lion")

class Elephant(Animal): 
    def __str__(self):
        return"Elephant"

    def Say(self):
        print("I Am a Elephant")

class Polar_Bear(Animal): 
    def __str__(self):
        return"Polar Bear"

    def Say(self):
        print("I Am a Polar Bear")

class Grizzly_Bear(Animal):
    def __str__(self):
        return"Grizzly Bear"
    
    def Say(self):
        print("I Am a Grizzly Bear")

class Moose(Animal):
    def __str__(self):
        return"Moose"
     
    def Say(self):
        print("I Am a Moose")

class Penguin(Animal):
    def __str__(self):
        return"Penguin"
     
    def Say(self):
        print("I Am a Penguin")
```
