```python
from typing import Type, Self 
from abc import ABC, abstractmethod
from operator import methodcaller
```

```python
class Firstname:
    def __init__(self, value):
        self.Fname = value

    def __str__(self):
        return f"Firstname: {self.Fname}"

class Middlename:
    def __init__(self, value):
        self.Mname = value

    def __str__(self):
        return f"Middlename: {self.Mname}"
    
class Lastname:
    def __init__(self, value):
        self.Lname = value

    def __str__(self):
        return f"Lastname: {self.Lname}"

class PrimaryEmail:
    def __init__(self, value):
        self.Pemail = value

    def __str__(self):
        return f"PrimaryEmail: {self.Pemail}"

class SecondaryEmail:
    def __init__(self, value):
        self.Semail = value

    def __str__(self):
        return f"SecondaryEmail: {self.Semail}"

class PrimaryPhone:
    def __init__(self, value):
        self.Pphone = value

    def __str__(self):
        return f"PrimaryPhone: {self.Pphone}"
    
class SecondaryPhone:
    def __init__(self, value):
        self.Sphone = value

    def __str__(self):
        return f"SecondaryPhone: {self.Sphone}"
```

```python
class Product:
    def __init__(self) -> None:
        self.__name="Customer Info"
        self.__components = []
        
    def add(self, component) -> None:
        print(f"The component {component} is being added now...")
        self.__components.append(component)
        
    def showComponents(self) -> None:
        print(f"There are {len(self.__components)} component(s) in the {self.__name} ")
        print(f"The components of {self.__name} are:\n")
        for item in self.__components:
            print(item)
```

```python
class Builder(ABC):
    @property
    @abstractmethod
    def product(self) -> None:#getResult method --> Army
        pass

    @abstractmethod
    def AddFirstName(self, str) -> None:
        pass

    @abstractmethod
    def AddMiddleName(self, str) -> None:
        pass

    @abstractmethod
    def AddLastName(self, str) -> None:
        pass
    
    @abstractmethod
    def AddPrimaryEmail(self, str) -> None:
        pass

    @abstractmethod
    def AddSecondaryEmail(self, str) -> None:
        pass

    @abstractmethod
    def AddPrimaryPhone(self, str) -> None:
        pass

    @abstractmethod
    def AddSecondaryPhone(self, str) -> None:
        pass
```

```python
class WebCustomerBuilder(Builder):
    def __init__(self) -> None:
        self.reset()
        
    def reset(self) -> None:
        print("We have our Web Customer Info \n")
        self.__product = Product()
        
    @property
    def product(self) ->Product:
        currentProduct=self.__product
        self.reset()#ready to start producing another product
        return currentProduct

    def AddFirstName(self, str) -> None:
        self.__product.add(Firstname(str))

    def AddMiddleName(self, str) -> None:
        self.__product.add(Middlename(str))

    def AddLastName(self, str) -> None:
        self.__product.add(Lastname(str))
    
    def AddPrimaryEmail(self, str) -> None:
        self.__product.add(PrimaryEmail(str))

    def AddSecondaryEmail(self, str) -> None:
        self.__product.add(SecondaryEmail(str))

    def AddPrimaryPhone(self, str) -> None:
        self.__product.add(PrimaryPhone(str))

    def AddSecondaryPhone(self, str) -> None:
        self.__product.add(SecondaryPhone(str))
    
```

```python
class MobileCustomerBuilder(Builder):
    
    def __init__(self) -> None:
        self.reset()
        
    def reset(self) -> None:
        print("We have our Mobile Customer Info \n")
        self.__product = Product()
        
    @property
    def product(self) ->Product:
        currentProduct=self.__product
        self.reset()#ready to start producing another product
        return currentProduct

    def AddFirstName(self, str) -> None:
        self.__product.add(Firstname(str))

    def AddMiddleName(self, str) -> None:
        self.__product.add(Middlename(str))

    def AddLastName(self, str) -> None:
        self.__product.add(Lastname(str))

    def AddPrimaryEmail(self, str) -> None:
        self.__product.add(PrimaryEmail(str))

    def AddSecondaryEmail(self, str) -> None:
        self.__product.add(SecondaryEmail(str))

    def AddPrimaryPhone(self, str) -> None:
        self.__product.add(PrimaryPhone(str))

    def AddSecondaryPhone(self, str) -> None:
        self.__product.add(SecondaryPhone(str))
    
```

```python
from typing import Type
class Director:
    def __init__(self) -> None:
        self.__builder = None

    @property
    def builder(self) -> Builder:
        return self.__builder

    @builder.setter
    def builder(self, builder: Type[Builder]) -> None:
        self.__builder=builder
        
    def AddDetails(self, Fname, Lname, Pemail, Pphone, Mname = None, Semail = None, Sphone = None)-> None:
        self.builder.AddFirstName(Fname)
        if Mname != None:
            self.builder.AddMiddleName(Mname)
        self.builder.AddLastName(Lname)
        self.builder.AddPrimaryEmail(Pemail)
        if Semail != None:
            self.builder.AddSecondaryEmail(Semail)
        self.builder.AddPrimaryPhone(Pphone)
        if Sphone != None:
            self.builder.AddSecondaryPhone(Sphone)
        
        self.builder.product.showComponents()
```

```python
def Customer():
    CustomerDirector=Director()
    
    WebBuilder = WebCustomerBuilder()
    CustomerDirector.builder=WebBuilder
    CustomerDirector.AddDetails("John","Forjd","John@email.com","7807807800","Henry","John@email.ca","5875875877")
    print('\n')

    MobileBuilder=MobileCustomerBuilder()
    CustomerDirector.builder=MobileBuilder
    CustomerDirector.AddDetails("Mary","petal","Mary@email.com","1234567890")
```
